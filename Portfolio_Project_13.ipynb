```python
data_1 = data.dropna()
data_1
```

```python
data_1["Date"] = pd.to_datetime(data_1["Date"], utc=True ,format="mixed")
data_1
```

```python
data_1.info()
```

```python
data_1['Price'] = data_1['Price'].str.replace("$","")
data_1['Price'] = data_1['Price'].str.replace(",","")
```

```python
data_1
```

```python
data_1.info()
```

```python
data_1['Price'] = pd.to_numeric(data_1['Price'])
```

```python
data_1.info()
```

```python
group_1 = data_1.groupby('Organisation')
group_1.value_counts()
```

```python
spending_analysis = group_1['Price'].sum().sort_values(ascending=False)
spending_analysis
```

```python
spending_df = spending_analysis.reset_index()
spending_df.columns = ['Organisation', 'Total_Spending']
```

```python
spending_df
```

```python
import plotly.express as px
```

```python
fig = px.treemap(spending_df,path=["Organisation"],values="Total_Spending",title='Total Space Mission Spending by Organization',
                 color='Total_Spending',
                 color_continuous_scale='Viridis')
fig.show()
```

```python
data_1['Country'] = data_1['Location'].str.split(',').str[-1].str.strip()
data_1['Country'].value_counts()
```

```python
data_1
```

```python
country_map = {
    'Russian Federation': 'Russia',
    'Kazakhstan': 'Russia',
    'New Mexico': 'USA',
    'Yellow Sea': 'China',
    'Shahrud Missile Test Site': 'Iran',
    'Pacific Ocean': 'USA'
}
data_1['Country'] = data_1['Country'].replace(country_map)
```

```python
country_launches = data_1['Country'].value_counts().reset_index()
country_launches.columns = ['Country', 'Launches']
fig = px.choropleth(country_launches,
                    locations="Country",
                    locationmode='country names',
                    color="Launches",
                    hover_name="Country",
                    title='Number of Space Launches by Country',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

```python
launch = px.sunburst(data_1, path=['Organisation', 'Mission_Status'], values='Price')
launch.show()
```

```python
data_1['Year'] = data_1['Date'].dt.year
data_1['Month'] = data_1['Date'].dt.month_name()
```

```python
launches_per_year = data_1.groupby(['Year', 'Organisation']).size().reset_index(name='Count')

fig = px.line(launches_per_year,
              x='Year',
              y='Count',
              color='Organisation',
              title='Number of Launches per Year by Organisation')
fig.show()
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
month_counts = data_1['Month'].value_counts().reset_index()
month_counts.columns = ['Month', 'Count']

fig = px.bar(month_counts,
             x='Month',
             y='Count',
             title='Most Popular Months for Space Launches',
             category_orders={'Month': month_order})
fig.show()
```

```python
data_1.to_csv('Space_Missions_Cleaned.csv', index=False)
```
